```python
import matplotlib.pyplot as plt
import numpy as np

grid=0
frame = 0

plt.rcParams['legend.fontsize'] = 12.5

# Generate sample data
x = np.linspace(-6, 10, 100)
mu_control, sigma_control = -1.1, 1.8
mu_treated, sigma_treated = 0.3, 1.8

y_control = np.exp(-0.5 * ((x - mu_control) / sigma_control) ** 2) / (sigma_control * np.sqrt(2 * np.pi))
y_treated = np.exp(-0.5 * ((x - mu_treated) / sigma_treated) ** 2) / (sigma_treated * np.sqrt(2 * np.pi))

# Create the plot
fig, ax = plt.subplots(3, 1, sharex=True)

y_ticks = [0.0, 0.1, 0.2]
ax[0].set_yticks(y_ticks)

ax[0].plot(x, y_control, label='$P (X |T=1, E=host)$', color='#1f77b4')
ax[0].plot(x, y_treated, label='$P (X |T=0, E=host)$', color='#ff7f0e')

# ax[0].set_xlabel('covariate')
ax[0].set_ylabel('density', fontsize=12)
ax[0].set_title('')
ax[0].grid(grid)
ax[0].legend(frameon=False)

ax[0].fill_between(x, y_control, alpha=0.25, color='#1f77b4')
ax[0].fill_between(x, y_treated, alpha=0.25, color='#ff7f0e')

ax[0].spines['right'].set_visible(False)
ax[0].spines['top'].set_visible(False)

########################################################################################################

# Set the y-axis ticks
ax[1].set_yticks(y_ticks)

ax[1].fill_between(x, y_control, alpha=0.25, color='#ff7f0e')
ax[1].fill_between(x, y_treated, alpha=0.25, color='#1f77b4')

ax[1].spines['right'].set_visible(False)
ax[1].spines['top'].set_visible(False)
 
ax[1].plot(x, y_control, label='$P (X| T=0, E=comp)$', color='#ff7f0e')
ax[1].plot(x, y_treated, label='$P (X| T=1, E=comp)$', color='#1f77b4')

# ax[1].set_xlabel('covariate')
ax[1].set_ylabel('density', fontsize=12)
ax[1].set_title('')
ax[1].grid(grid)
ax[1].legend(frameon=False)

########################################################################################################

# Set the y-axis ticks

ax[2].plot(x, y_control, label='$P (X | T=1, E=twin)$', color='#1f77b4')
ax[2].plot(x, y_treated, label='$P (X | T=0, E=twin)$', color='#ff7f0e')

ax[2].set_xlabel('covariate', fontsize=12)
ax[2].set_ylabel('density', fontsize=12)
ax[2].set_title('')
ax[2].grid(grid)

# Set the y-axis ticks
ax[2].set_yticks(y_ticks)

ax[2].fill_between(x, y_control, alpha=0.25, color='#1f77b4')
ax[2].fill_between(x, y_treated, alpha=0.25, color='#ff7f0e')

ax[2].spines['right'].set_visible(False)
ax[2].spines['top'].set_visible(False)
ax[2].legend(frameon=frame)

plt.tight_layout()
plt.savefig('/Users/lucile/causal_info_gain/plots/exp_plots.pdf', dpi=300)
plt.show()
```
